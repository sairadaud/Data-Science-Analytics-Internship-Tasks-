```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Load the dataset
iris = sns.load_dataset("iris")
```

In the above cell we loaded the Iris dataset directly from the Seaborn library.

This dataset is preloaded in Seaborn and does not require any external file.

It contains measurements of iris flowers and is commonly used for:

Classification problems
Data visualization practice
Statistical analysis

```python
#Initial Dataset Exploaration
iris.shape
iris.columns
iris.head()

#Explanation:
#shape shows the number of rows and columns in the dataset
#columns displays all feature names
#head() shows the first 5 rows of the dataset
```

Dataset Overview
The Iris dataset contains 150 samples of iris flowers belonging to three different species:

Setosa
Versicolor
Virginica

Each sample includes 4 key features:

Sepal Length
Sepal Width
Petal Length
Petal Width

These features represent physical measurements of flowers and are used to distinguish between species.

This dataset is considered a balanced dataset, meaning each species has an equal number of samples, which makes it ideal for classification tasks.

```python
#Statistical Summary

iris.describe()

# Explanation:

# The describe() function provides statistical insights about numerical features such as:

# Mean (average value)
# Standard deviation (spread of data)
# Minimum and maximum values
# Quartiles (25%, 50%, 75%)
```

Observations from Summary Statistics:
Sepal length values range from small to relatively large variations
Sepal width has less variation compared to other features
Petal length shows the highest variation among all features
Mean values help understand the central tendency of each feature
The dataset shows natural variation across flower species

Observations from Dataset Structure

From initial analysis, we observe:

The dataset contains 150 rows and 5 columns
There are no missing values, so no data cleaning is required
All columns are well-structured and complete
The dataset is balanced across the three species

This makes the dataset clean and suitable for visualization and modeling.

```python

#Scatter Plot: Petal Length vs Petal Width

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=iris,
    x="petal_length",
    y="petal_width",
    hue="species"
)

plt.title("Petal Length vs Petal Width")
plt.show()


#Explanation:

# This scatter plot shows the relationship between petal length 
# and petal width, with different colors representing different species.
```

Observations:
There is a clear separation between species
Setosa is easily distinguishable from the other two species
Versicolor and Virginica show some overlap but still form visible clusters
Petal features are strongly correlated
These features are very useful for classification tasks

```python
#Histogram: Distribution of Sepal Length

plt.figure(figsize=(8,6))

sns.histplot(
    iris["sepal_length"],
    bins=10,
    kde=True
)

plt.title("Distribution of Sepal Length")
plt.show()


#Explanation:

# A histogram helps us understand the distribution of a single numerical variable.
# The KDE curve shows the smooth distribution trend.
```

Observations:
Most values are concentrated in the mid-range
The distribution is smooth and fairly balanced
There is no strong skewness in the data
Sepal length follows a natural distribution pattern

```python

#Box Plot: Sepal Length by Species

plt.figure(figsize=(8,6))

sns.boxplot(
    data=iris,
    x="species",
    y="sepal_length"
)

plt.title("Sepal Length by Species")
plt.show()


#Explanation:

# A box plot is used to understand:

# Median values
# Data spread
# Presence of outliers
# Differences between categories


```

Observations:
Each species shows different sepal length distribution
Virginica generally has higher sepal length values
Setosa has lower and more compact values
Versicolor lies between the two
There is some overlap, but clear patterns still exist
